## BÚSQUEDA DE VALORES FALTANTES

```python
# Función para encontrar valores faltantes en un DataFrame
def buscar_valores_faltantes(df):
    valores_faltantes = df.isnull().sum()
    return valores_faltantes[valores_faltantes > 0]

# Iterar sobre cada hoja y buscar valores faltantes
valores_faltantes_dict = {}
for hoja, df in internet.items():
    valores_faltantes = buscar_valores_faltantes(df)
    if not valores_faltantes.empty:
        valores_faltantes_dict[hoja] = valores_faltantes

# Mostrar resultados
for hoja, valores_faltantes in valores_faltantes_dict.items():
    print(f'Hoja: {hoja}')
    print(valores_faltantes)
    print('-' * 40)
```

## BÚSQUEDA DE VALORES ATÍPICOS/EXTREMOS U OUTLIERS

```python
# Se observan la cantidad total de outliers por hoja y columna
summary = []

for nombre_hoja, df in internet.items():
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for columna in num_cols:
        Q1 = df[columna].quantile(0.25)
        Q3 = df[columna].quantile(0.75)
        IQR = Q3 - Q1
        num_outliers = df[(df[columna] < (Q1 - 1.5 * IQR)) | (df[columna] > (Q3 + 1.5 * IQR))].shape[0]
        summary.append({
            'Hoja': nombre_hoja,
            'Columna': columna,
            'Num_outliers': num_outliers
        })

summary_df = pd.DataFrame(summary)
summary_df.tail(25)
```

- #### Gráfica Penetración-totales: Se visualiza la cantidad de accesos por cada 100 hogares y 100 habitantes a lo largo de los meses por un periodo de 10 años. 

```python
# Función para expandir los periodos a rangos de fechas
def expandir_periodo_a_fechas(row):
    
    # Se crea un diccionario para mapear las abreviaturas de los meses a números
    meses = {
        'Ene': 1, 'Feb': 2, 'Mar': 3, 'Abr': 4, 'May': 5, 'Jun': 6,
        'Jul': 7, 'Ago': 8, 'Sept': 9, 'Oct': 10, 'Nov': 11, 'Dic': 12
    }
    
    # Se obtienen los valores de las columnas relevantes del DataFrame
    periodo = row['Periodo']
    accesos_hogares = row['Accesos por cada 100 hogares']
    accesos_habitantes = row['Accesos por cada 100 hab']
    
    # Se separa el periodo en rango de meses y año
    rango, año = periodo.split()
    # Separar el rango de meses en mes de inicio y mes de fin
    mes_inicio, mes_fin = rango.split('-')
    
    # Se convierte el mes de inicio y el año en una fecha de inicio
    fecha_inicio = pd.to_datetime(f'{meses[mes_inicio]} 01 {año}', format='%m %d %Y')
    # Se convierte el mes de fin y el año en una fecha de fin, ajustando al último día del mes
    fecha_fin = pd.to_datetime(f'{meses[mes_fin]} 01 {año}', format='%m %d %Y') + pd.DateOffset(months=1) - pd.DateOffset(days=1)
    
    # Se crea un rango de fechas mensuales entre fecha_inicio y fecha_fin
    fechas = pd.date_range(fecha_inicio, fecha_fin, freq='ME')
    
    # Se devuelve un DataFrame con las fechas y los accesos promedio distribuidos a lo largo de estas fechas
    return pd.DataFrame({
        'fecha': fechas, 
        'Accesos por cada 100 hogares': [accesos_hogares/len(fechas)]*len(fechas),
        'Accesos por cada 100 habitantes': [accesos_habitantes/len(fechas)]*len(fechas)
    })

# Se aplica la función para expandir los periodos en cada fila del DataFrame original
df_expandido = pd.concat([expandir_periodo_a_fechas(row) for _, row in Penetracion_totales.iterrows()], ignore_index=True)
```

```python
# Crear el gráfico de líneas para ambas series de datos
plt.figure(figsize=(12, 6))

sns.lineplot(x='fecha', y='Accesos por cada 100 hogares', data=df_expandido, marker='o', label='Accesos por cada 100 hogares')
sns.lineplot(x='fecha', y='Accesos por cada 100 habitantes', data=df_expandido, marker='o', label='Accesos por cada 100 habitantes')

plt.title('Accesos a lo Largo del Tiempo (por meses)')
plt.xlabel('Fecha')
plt.ylabel('Accesos')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

CONCLUSIÓN DE LA GRÁFICA: Se deduce que hay un mayor crecimiento en lo que respecta al acceso por hogares a comparación del acceso por habitantes.
